```python
import wave
wav_obj = wave.open('/content/gdrive/MyDrive/Principles of DataScience/Data Set/song.wav', 'rb')
```

```python
sample_freq = wav_obj.getframerate()
sample_freq
```

```python
n_samples = wav_obj.getnframes()
n_samples
```

```python
t_audio = n_samples/sample_freq
t_audio
```

```python
n_channels = wav_obj.getnchannels()
n_channels
```

```python
signal_wave = wav_obj.readframes(n_samples)
```

```python
 
import numpy as np
signal_array = np.frombuffer(signal_wave, dtype=np.int16)
```

```python
l_channel = signal_array[0::2]
r_channel = signal_array[1::2]
```

```python
times = np.linspace(0, n_samples/sample_freq, num=n_samples)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 5))
plt.plot(times, l_channel)
plt.title('Left Channel')
plt.ylabel('Signal Value')
plt.xlabel('Time (s)')
plt.xlim(0, t_audio)
plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.specgram(l_channel, Fs=sample_freq, vmin=-20, vmax=50)
plt.title('Left Channel')
plt.ylabel('Frequency (Hz)')
plt.xlabel('Time (s)')
plt.xlim(0, t_audio)
plt.colorbar()
plt.show()
```

```python
import cv2

# read video from file
cap = cv2.VideoCapture('/content/gdrive/MyDrive/Principles of DataScience/Data Set/file_example_MP4_480_1_5MG.mp4')
```

```python
video_fps = cap.get(cv2.CAP_PROP_FPS),
total_frames = cap.get(cv2.CAP_PROP_FRAME_COUNT)
height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)

print(f"Frame Per second: {video_fps } \nTotal Frames: {total_frames} \n Height: {height} \nWidth: {width}")
```

```python
from google.colab.patches import cv2_imshow
ret, frame = cap.read()
# show frame
cv2_imshow(frame)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
from google.colab.patches import cv2_imshow
while True:
    ret, frame = cap.read()
    if not ret: break # break if no next frame
    
    cv2_imshow(frame) # show frame
    
    if cv2.waitKey(1) & 0xFF == ord('q'): # on press of q break
        break
        
# release and destroy windows
cap.release()
cv2.destroyAllWindows()
```
